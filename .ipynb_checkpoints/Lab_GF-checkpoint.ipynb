```python
import pandas as pd
```

```python
df1 = pd.read_csv('files_for_lab/csv_files/file1.csv')
df2 = pd.read_csv('files_for_lab/csv_files/file2.csv')
df3 = pd.read_csv('files_for_lab/csv_files/file3.csv')
```

```python
df1.shape
```

```python
df2.shape
```

```python
df3.shape
```

```python
df1.columns = [col_name.lower().replace(' ', '_').replace('state','st') for col_name in df1.columns]
df1.columns
```

```python
df2.columns = [col_name.lower().replace(' ', '_').replace('state','st') for col_name in df2.columns]
df2.columns
```

```python
df3.columns = [col_name.lower().replace(' ', '_').replace('state','st') for col_name in df3.columns]
df3.columns
```

```python
df1_std=df1
df2_std=df2[df1.columns]
df3_std=df3[df1.columns]
```

```python
data = pd.concat([df1_std, df2_std,df3_std])
```

```python
data.head()
```

```python
data = data.rename(columns={'customer':'id',
                            'customer_lifetime_value':'clvalue',
                            'monthly_premium_auto':'m_premium',
                           'number_of_open_complaints':'open_comp'})
```

```python
# Which columns are numerical?
numericals=data.select_dtypes("number")
numericals
```

```python
# Which columns are categorical?
categorical=data.select_dtypes("object")
categorical
```

```python
# Delete the column education and the number of open complaints from the dataframe.
data=data.drop(['education','open_comp'],axis=1)
```

```python
# Drop rows where all cells are NaNs
data=data.dropna(how='all')
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.
data['clvalue']=data['clvalue'].str.replace('%','')
data['clvalue']=data['clvalue'].apply(lambda x: x if pd.isna(x) else int(float(x)*100))
data['clvalue']
```

```python
data.clvalue.apply(type)
```

```python
data[data.duplicated()]
```

```python
# Check for duplicate rows in the data and remove if any.
data=data.drop_duplicates()
data
```

```python
# Filter out the data for customers who have an income of 0 or less.
data=data[data['income']>0]
data
```
